```python
import numpy as np

# Function to remove outliers using IQR
def remove_outliers(dataset, column):
    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)
    IQR = Q3 - Q1
    filter = (dataset[column] >= Q1 - 1.5 * IQR) & (dataset[column] <= Q3 + 1.5 * IQR)
    return dataset.loc[filter]

# Apply the function to each column (adjust as needed)
for column in dataset.select_dtypes(include=[np.number]).columns:
    df = remove_outliers(dataset, column)
```

```python

for col in dataset.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=dataset[col])
    plt.title(f'Box plot of {col}')
    plt.show()
```
